```python
from IPython.display import display
import ipywidgets as widgets

#tweets = [...]  # your loaded tweet objects
index = 0
#labels = []

def get_next_tweet():
    global index
    if index >= len(tweets):
        return "No more tweets."
    txt = tweets[index]["text"]
    return txt

def label_and_next(value):
    global index
    labels.append(value)
    index += 1
    tweet_box.value = get_next_tweet()
```

```python
tweet_box = widgets.HTML(value=get_next_tweet())

button_relevant = widgets.Button(description="Relevant (1)", button_style='success')
button_not = widgets.Button(description="Not Relevant (0)", button_style='danger')

def on_rel_clicked(b):
    label_and_next(1)

def on_not_clicked(b):
    label_and_next(0)

button_relevant.on_click(on_rel_clicked)
button_not.on_click(on_not_clicked)

display(tweet_box, button_relevant, button_not)
```
