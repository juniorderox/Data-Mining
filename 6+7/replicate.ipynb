```python
input_filename = "tweets.json"
labels_filename = "python_classes.json"
replicable_dataset = "replicable_dataset.json"
```

```python
import json
import os


# --- load all tweets ---
with open(input_filename) as inf:
    data = json.load(inf)       # load full JSON
    tweets = data["data"]       # list of tweet objects

# --- load existing labels ---
labels = []
if os.path.exists(labels_filename):
    with open(labels_filename) as inf:
        labels = json.load(inf)

print("Loaded tweets:", len(tweets))
print("Loaded labels:", len(labels))
```

```python
dataset = [(tweet['id'], label) for tweet, label in zip(tweets, labels)]
```

```python
# --- save replicable dataset ---
with open(replicable_dataset, 'w') as outf:
    json.dump(dataset, outf)
```
